```python
data = pd.read_csv( '/content/drive/MyDrive/repos/house_sales_prediction/data/kc_house_data.csv' )
```

```python
# 1. Crie uma nova coluna chamada: “dormitory_type”
    # - Se o valor da coluna “bedrooms” for igual à 1 => ‘studio’
    # - Se o valor da coluna “bedrooms” for igual a 2 => ‘apartament’
    # - Se o valor da coluna “bedrooms” for maior que 2 => ‘house’

# - Se o valor da coluna “bedrooms” for igual à 1 => ‘studio’
data.loc[data['bedrooms'] == 1, 'dormitory_type'] = 'studio'

# - Se o valor da coluna “bedrooms” for igual a 2 => ‘apartament’
data.loc[data['bedrooms'] == 2, 'dormitory_type'] = 'apartament'

# - Se o valor da coluna “bedrooms” for maior que 2 => ‘house’
data.loc[data['bedrooms'] > 2, 'dormitory_type'] = 'house'
```

```python
# Verificar se todos as linhas foram preenchidas

aux1 = data[['dormitory_type', 'bedrooms']].groupby( 'dormitory_type' ).sum().reset_index()
aux1
```

```python
# Faça um gráfico de barras que represente a soma dos preços pelo número de quartos.

by_bedrooms = data[['price', 'bedrooms']].groupby( 'bedrooms' ).sum().reset_index()
plt.figure( figsize=( 20, 12 ) )
plt.bar( by_bedrooms['bedrooms'], by_bedrooms['price'] )
```

```python
# Faça um gráfico de barras que represente a média dos preços pelo tipo dos dormitórios.

by_dorm_type = data[['price', 'dormitory_type']].groupby( 'dormitory_type' ).mean().reset_index()
plt.figure( figsize=( 20, 12 ) )
plt.bar( by_dorm_type['dormitory_type'], by_dorm_type['price'] )
```

```python
# Faça um gráfico de linha que mostre a evolução da média dos preços pelo ano da reforma dos imóveis, a partir do ano de 1930.

#data.loc( ['yr_renovated'] >= 1930, 'yr_renovated' )
by_yr = data[['price'], ['yr_renovated' >= 1930 ].groupby( 'yr_renovated' ).mean().reset_index()
plt.plot( by_yr['price'], by_yr['yr_renovated'] )
```

```python
# 6. Faça um tabela que mostre a média dos preços por ano de construção e tipo de dormitórios dos imóveis.

aux1 = data[['price', 'yr_built', 'dormitory_type']].groupby( 'yr_built', 'dormitory_type' ).mean().reset_index()
aux1.head()
```

```python
# 7. Crie um Dashboard com os gráficos das questões 02, 03, 04 ( Dashboard: 1 Linha e 2 colunas )
# 8. Crie um Dashboard com os gráficos das perguntas 02, 04 ( Dashboard: 2 colunas )
# 9. Crie um Dashboard com os gráficos das perguntas 03, 05 ( Dashboard: 2 Linhas )

from matplotlib import gridspec


fig = plt.figure( figsize=( 20, 12 ) )
specs = gridspec.GridSpec( ncols=2, nrows=2, figure=fig )


ax1 = fig.add_subplot( specs[0, :] ) # First Row
ax2 = fig.add_subplot( specs[1, 0] ) # Second Rows - Firt Column
ax3 = fig.add_subplot( specs[1, 1] ) # Second Rows - Second Column


# Fisrt Graph
by_bedrooms = data[['price', 'bedrooms']].groupby( 'bedrooms' ).sum().reset_index()
ax1.bar( by_bedrooms['bedrooms'], by_bedrooms['price'] )

# Second Graph
by_built = data[['price','yr_built']].groupby( 'yr_built' ).mean().reset_index()
ax2.plot( by_built['yr_built'], by_built['price'])

# Third Graph
by_dorm_type = data[['price', 'dormitory_type']].groupby( 'dormitory_type' ).mean().reset_index()
ax3.bar( by_dorm_type['dormitory_type'], by_dorm_type['price'] )
```
